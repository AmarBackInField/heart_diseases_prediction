```python
import pandas as pd
```

```python
df = pd.read_csv('data\cleaned_heart_diseases.csv')
df.head()
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
X=df.drop('target',axis=1)
y=df['target']
```

```python
from sklearn.preprocessing import StandardScaler # HAndling Feature Scaling
# from sklearn.preprocessing import OrdinalEncoder # Ordinal Encoding -- In this dataset no categarical column present
## pipelines
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
numerical_cols=X.select_dtypes(include=['int64','float64']).columns
```

```python
numerical_cols
```

```python
## Numerical Pipeline
num_pipeline=Pipeline(
    steps=[
    ('scaler',StandardScaler())
    ]

)


preprocessor=ColumnTransformer([
('num_pipeline',num_pipeline,numerical_cols),
])
```

```python
## Train test split

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=30)
```

```python
X_train=pd.DataFrame(preprocessor.fit_transform(X_train),columns=preprocessor.get_feature_names_out())
X_test=pd.DataFrame(preprocessor.transform(X_test),columns=preprocessor.get_feature_names_out())
```

```python
X_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

models = {
    'DecisionTreeClassifier': DecisionTreeClassifier(),
    'RandomForestClassifier': RandomForestClassifier(),
    'GaussianNB':GaussianNB(),
    'GradientBoostingClassifier': GradientBoostingClassifier()
    
}

trained_model_list = []
model_names = []
accuracy_scores = []

def evaluate_model(y_true, y_pred):
    # Your evaluation code goes here
    # For example:
    confusion_mt = confusion_matrix(y_true, y_pred)
    classification_rt = classification_report(y_true, y_pred)
    accuracy_sc = accuracy_score(y_true, y_pred)
    return confusion_mt, classification_rt, accuracy_sc

for model_name, model in models.items():
    model.fit(X_train, y_train)

    # Make Predictions
    y_pred = model.predict(X_test)

    confusion_mt, classification_rt, accuracy_sc = evaluate_model(y_test, y_pred)

    print(model_name)
    model_names.append(model_name)

    print('\n Model Training Performance')
    print("confusion_matrix:", confusion_mt)
    print("classification_report:", classification_rt)
    print("accuracy_score", accuracy_sc * 100)

    accuracy_scores.append(accuracy_sc)

    print('=' * 35)
    print('\n')
```
